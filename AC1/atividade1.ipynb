## Parte 1 - Simulao de uma fonte sonora
Nessa etapa, ser realizada uma simulao do sinal bineural obtido a partir de uma fonte sonora em um dos pontos medidos.  
O objetivo  obter uma representao grfica e sonora do sinal observado, simulando o que seria observado pelo canal auditivo humano.



Assumiremos o canal auditivo humano como um sistema LIT, o qual  completamente caracterizado por sua resposta ao impulso (Arquivos em Medies/SEC_*.csv). Analisaremos o audio udios/4Seasons_Alto.wav no Ponto 1.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import soundfile as sf
import IPython.display as ipd
import pandas as pd
```

### 1.1 - Leitura e Setup

```python
h1 = np.genfromtxt('Medies/SEC_1.csv', delimiter=',') #Resposta ao impulso do sistema no ponto 1
audio, fs = sf.read('udios/4Seasons_Alto.wav') #Leitura do udio

#Normlizao da resposta ao impulso
h1 = (h1/np.ptp(h1,axis=0))
audio = (audio/np.ptp(audio,axis=0))

#Reduzindo o tamanho do audio para um minuto
audio = audio[0:fs*60]

#Vetor tempo do udio. 1/fs  perodo de uma amostra
t = np.arange(len(audio)) * 1/fs

#Diviso da resposta ao impulso em lado direito e esquerdo
h1dir = h1[:,0] #Canal direito
h1esq = h1[:,1] #Canal esquerdo
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 8))

#Plot do audio original
axs[0,1].plot(t, audio)
```
