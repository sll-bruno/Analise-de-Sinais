## Parte 1 - Simulação de uma fonte sonora
Nessa etapa, será realizada uma simulação do sinal bineural obtido a partir de uma fonte sonora em um dos pontos medidos.  
O objetivo é obter uma representação gráfica e sonora do sinal observado, simulando o que seria observado pelo canal auditivo humano.



Assumiremos o canal auditivo humano como um sistema LIT, o qual é completamente caracterizado por sua resposta ao impulso (Arquivos em Medições/SEC_*.csv). Analisaremos o audio Áudios/4Seasons_Alto.wav no Ponto 1.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import soundfile as sf
import IPython.display as ipd
import pandas as pd
```

### 1.1 - Leitura e Setup

```python
h1 = np.genfromtxt('Medições/SEC_1.csv', delimiter=',') #Resposta ao impulso do sistema no ponto 1
audio, fs = sf.read('Áudios/4Seasons_Alto.wav') #Leitura do áudio

#Normlização da resposta ao impulso
h1 = (h1/np.ptp(h1,axis=0))
audio = (audio/np.ptp(audio,axis=0))

#Reduzindo o tamanho do audio para um minuto
audio = audio[0:fs*60]

#Vetor tempo do áudio. 1/fs é período de uma amostra
t = np.arange(len(audio)) * 1/fs

#Divisão da resposta ao impulso em lado direito e esquerdo
h1dir = h1[:,0] #Canal direito
h1esq = h1[:,1] #Canal esquerdo
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10, 8))

#Plot do audio original
axs[0,1].plot(t, audio)
```
